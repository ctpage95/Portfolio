```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read in the data and check to confirm data were read in correctly

```python
storms = pd.read_csv('storms.csv')
```

```python
storms.head()
```

### Create a scatterplot between wind and pressure, with color being used to distinguish the type of storm.  Also draw the four regression lines (one for each type of storm)

```python
sns.lmplot(data=storms, x="wind",  y="pressure",hue='type', height=12)
```

### Repeat the plot from but use facets.  

```python
sns.lmplot(data=storms, x="wind",  y="pressure", col= 'type', hue='type', height=12, col_wrap=2)
```

### Plot the path of each storm in the data.  Use color to distinguish the storms from one another, and use facetting to plot each year in its own panel.  


```python
g = sns.FacetGrid(storms, col="year",hue='type')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "long", "lat").add_legend()
```
